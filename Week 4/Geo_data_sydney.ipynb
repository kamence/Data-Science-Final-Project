```python
#getting the table from the below URL
url = 'https://www.geonames.org/postal-codes/AU/NSW/new-south-wales.html'
tables = pd.read_html(url)
```

```python
#we choose the 2nd table from the webpage
geo_code = tables[2]
```

```python
#we want the name of the columns
geo_code.columns
```

```python
#display table sydney
geo_code
```

```python
geo_code = geo_code['Code'].to_frame()
geo_code
```

```python
geo_code.dropna(subset=['Code'], inplace=True)
geo_code
```

```python
 geocode = geo_code.reset_index()
```

```python
geocode.columns
```

```python
geocode.drop(columns=['index'], inplace=True)
geocode
```

```python
geocode["Lattitude"] = ""
```

```python
geocode["Longitude"] = ""
geocode
```

```python
new_geocode = pd.DataFrame({'Code':geocode['Code'].iloc[::2].values, 'Lattitude':geocode['Code'].iloc[1::2].values})
new_geocode
```

```python
new_geocode[['Lattitude','Longitude']] = new_geocode['Lattitude'].str.split('/',expand=True)
new_geocode
```

```python
new_geocode.rename(columns={"Code": "Postal Code"}, inplace=True)
new_geocode
```

```python
new_geocode.to_csv('sydney_geocode.csv', index=False)  
print("Table saved as csv file!")
```
